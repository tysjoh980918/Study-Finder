```python
%load_ext ipython_sparql_pandas
import pandas as pd
```

```python
%%sparql http://localhost:7200/repositories/KandD_Project -q -s universities

        
BASE <http://example.com/base/>
PREFIX mapper: <http://www.ontotext.com/mapper/>
PREFIX ex: <http://www.semanticweb.org/user/ontologies/Project/>
PREFIX rdfs: <http://www.w3.org/2000/01/rdf-schema#>
PREFIX xsd: <http://www.w3.org/2001/XMLSchema#>

SELECT ?programme ?university ?descriptionLabel
WHERE { 
	?programme a ex:Programme ;
            ex:hasAFieldOfStudy ?fieldofstudy;
            ex:hasATypeOfProgrammes ?programmeType;
    		ex:isOfferedBy ?university;
      		ex:isFullTime ?fullTime;
        	ex:isPartTime ?partTime; 
    		a ?class; 
    		ex:hasDescription ?description;
      		ex:isTaughtIn ?language.
    ?university ex:isLocatedAt ?city;
                ex:hasATypeOfInstitution ?institutionType .
    ?description rdfs:label ?descriptionLabel .
    FILTER(?programmeType = ex:Master)
    FILTER(?institutionType = ex:WO)
    FILTER(?language = ex:en)
    FILTER(?partTime = true)
    FILTER(?class = ex:Technology)
    FILTER(?city = ex:AMSTERDAM)
} limit 100 
```

```python
# This table will show the available programs with the universities and a description of what the study offers
## 
```

```python
universities['programme'] = universities['programme'].replace(r'http://www.semanticweb.org/user/ontologies/Project/', '', regex=True)
universities['university'] = universities['university'].replace(r'http://www.semanticweb.org/user/ontologies/Project/', '', regex=True)
universities['descriptionLabel'] = universities['descriptionLabel'].replace(r'http://www.semanticweb.org/user/ontologies/Project/', '', regex=True)

df = pd.DataFrame(list(zip(universities['programme'], universities['university'], universities['descriptionLabel'])),
                  columns = ['Programme','University','Description'])
df
```
