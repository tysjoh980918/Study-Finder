```python
from rdflib import Graph, RDF, Namespace, Literal, URIRef
from SPARQLWrapper import SPARQLWrapper, JSON
```

```python
latitude = []
longitude = []
university = []



sparql = SPARQLWrapper("http://dbpedia.org/sparql")
sparql.setQuery(
    """
PREFIX dbc: <http://dbpedia.org/resource/Category:>
PREFIX db: <http://dbpedia.org/>
PREFIX ca: <http://complyadvantage.com/>
PREFIX dbo: <http://dbpedia.org/ontology/>
PREFIX category:<http://dbpedia.org/resource/Category:>
PREFIX dcterms:<http://purl.org/dc/terms/> 
PREFIX geo:<http://www.w3.org/2003/01/geo/wgs84_pos#>
PREFIX dbpedia-owl:<http://dbpedia.org/ontology/>
PREFIX dbpedia: <http://dbpedia.org/resource/>

    SELECT ?university ?lat ?long
    WHERE {
    
{  
?university a dbc:Universities_in_the_Netherlands;
  dbpedia-owl:country dbpedia:Netherlands.
} UNION { 
  ?university dcterms:subject category:Universities_in_the_Netherlands.
}

?university
  geo:long ?long;
  geo:lat ?lat .

    }
""")

sparql.setReturnFormat(JSON)
results = sparql.query().convert()
for result in results ["results"]["bindings"]:
    latitude.append(float(result["lat"]["value"]))
     
for result2 in results ["results"]["bindings"]:
    longitude.append(float(result2["long"]["value"]))
    
for result3 in results ["results"]["bindings"]:
    university.append(result3["university"]["value"])


```

```python
sum(latitude)/(len(latitude))
sum(longitude)/(len(longitude))

middle = [sum(latitude)/(len(latitude)),sum(longitude)/(len(longitude))]
m = folium.Map(location=middle, zoom_start=7)
for i in range(len(latitude)):
   folium.CircleMarker(
        location = [latitude[i], longitude[i]],
        radius= 10,
        popup= [university[i].split("/")[-1]],
        color="#3186cc",
        fill=True,
        fill_color="#3186cc",
    ).add_to(m)

m
```
